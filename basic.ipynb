#### 02.2 결측치 비율 확인

```python
# 결측치 비율 확인 함수

def missing_summary(df):
    missing_df = pd.DataFrame({
        "column": df.columns,
        "missing_count": df.isna().sum().values,
        "missing_rate": (df.isna().mean().values * 100).round(2),
        "non_missing_count": df.notna().sum().values,
        "dtype": df.dtypes.astype(str).values,
        "unique_count": df.nunique(dropna=True).values
    })
    
    missing_df = missing_df.sort_values(
        by=["missing_rate", "missing_count"],
        ascending=False
    ).reset_index(drop=True)
    
    return missing_df
```

```python
museum_missing = missing_summary(museum_basic)
art_missing = missing_summary(art_basic)

print("박물관 결측치 요약")
display(museum_missing)

print("미술관 결측치 요약")
display(art_missing)
```

```python
print("박물관 결측치가 있는 컬럼")
display(museum_missing[museum_missing["missing_count"] > 0])

print("미술관 결측치가 있는 컬럼")
display(art_missing[art_missing["missing_count"] > 0])
```

```python
# 결측치 비율 구간별 컬럼 확인

def check_missing_groups(missing_df, dataset_name):
    print(f"[{dataset_name}] 결측치 70% 이상 컬럼")
    display(missing_df[missing_df["missing_rate"] >= 70])
    
    print(f"[{dataset_name}] 결측치 30% 이상 70% 미만 컬럼")
    display(missing_df[
        (missing_df["missing_rate"] >= 30) & 
        (missing_df["missing_rate"] < 70)
    ])
    
    print(f"[{dataset_name}] 결측치 0% 초과 30% 미만 컬럼")
    display(missing_df[
        (missing_df["missing_rate"] > 0) & 
        (missing_df["missing_rate"] < 30)
    ])


check_missing_groups(museum_missing, "박물관")
check_missing_groups(art_missing, "미술관")
```

#### 02.3 결측치 처리 기준 정리

```python
# 결측치 비율에 따른 1차 처리 기준 생성 함수

def add_missing_decision(missing_df):
    missing_df = missing_df.copy()
    
    def decide(row):
        rate = row["missing_rate"]
        
        if rate == 100:
            return "제거 후보"
        elif rate >= 70:
            return "보류/제거 후보"
        elif rate >= 30:
            return "의미 검토 필요"
        elif rate > 0:
            return "결측 처리 후 사용 가능"
        else:
            return "유지 가능"
    
    missing_df["missing_decision"] = missing_df.apply(decide, axis=1)
    
    return missing_df


museum_missing_decision = add_missing_decision(museum_missing)
art_missing_decision = add_missing_decision(art_missing)

print("박물관 결측치 처리 기준")
display(museum_missing_decision)

print("미술관 결측치 처리 기준")
display(art_missing_decision)
```

```python
print("[박물관 결측치 처리 기준별 컬럼 수]")
display(museum_missing_decision["missing_decision"].value_counts())

print("[미술관 결측치 처리 기준별 컬럼 수]")
display(art_missing_decision["missing_decision"].value_counts())
```

```python
print("[박물관 제거/보류 후보]")
display(
    museum_missing_decision[
        museum_missing_decision["missing_decision"].isin(["제거 후보", "보류/제거 후보"])
    ]
)

print("[미술관 제거/보류 후보]")
display(
    art_missing_decision[
        art_missing_decision["missing_decision"].isin(["제거 후보", "보류/제거 후보"])
    ]
)
```

```python
# 숫자형 변환 전 점검 함수
# 쉼표를 제거한 뒤 numeric 변환이 가능한 비율을 확인한다.
# 이 셀에서는 실제 변환은 하지 않는다.

def numeric_conversion_check(df, cols):
    rows = []
    
    for col in cols:
        s = df[col]
        non_missing_count = s.notna().sum()
        
        cleaned = (
            s.astype("string")
             .str.strip()
             .str.replace(",", "", regex=False)
        )
        
        converted = pd.to_numeric(cleaned, errors="coerce")
        numeric_count = converted.notna().sum()
        
        rows.append({
            "column": col,
            "dtype_before": str(s.dtype),
            "non_missing_count": non_missing_count,
            "numeric_convertible_count": numeric_count,
            "numeric_convertible_rate": round((numeric_count / non_missing_count) * 100, 2) if non_missing_count > 0 else 0,
            "sample_values": " | ".join(s.dropna().astype(str).unique()[:5])
        })
    
    return pd.DataFrame(rows)


museum_numeric_check = numeric_conversion_check(museum_typed, museum_numeric_cols)
art_numeric_check = numeric_conversion_check(art_typed, art_numeric_cols)

print("[박물관 숫자형 변환 후보 점검]")
display(museum_numeric_check)

print("[미술관 숫자형 변환 후보 점검]")
display(art_numeric_check)
```

```python
# 95% 이상 변환 가능하지만 일부 비숫자 값이 섞인 컬럼 확인
# 실제 변환 전에 어떤 값이 numeric 변환을 막는지 확인한다.

def check_non_numeric_values(df, check_df, dataset_name, lower_rate=95, upper_rate=100):
    target_cols = check_df[
        (check_df["numeric_convertible_rate"] >= lower_rate) &
        (check_df["numeric_convertible_rate"] < upper_rate)
    ]["column"].tolist()
    
    rows = []
    
    for col in target_cols:
        s = df[col]
        
        cleaned = (
            s.astype("string")
             .str.strip()
             .str.replace(",", "", regex=False)
        )
        
        converted = pd.to_numeric(cleaned, errors="coerce")
        
        # 원래 값은 있는데 numeric 변환 결과가 NaN인 값
        non_numeric_values = s[
            s.notna() & converted.isna()
        ].astype(str).unique()
        
        rows.append({
            "dataset": dataset_name,
            "column": col,
            "dtype_before": str(s.dtype),
            "non_numeric_count": len(non_numeric_values),
            "non_numeric_values": " | ".join(non_numeric_values[:10])
        })
    
    return pd.DataFrame(rows)


museum_ambiguous_numeric = check_non_numeric_values(
    museum_typed,
    museum_numeric_check,
    "박물관"
)

art_ambiguous_numeric = check_non_numeric_values(
    art_typed,
    art_numeric_check,
    "미술관"
)

print("[박물관] 95% 이상 100% 미만 숫자형 변환 후보의 비숫자 값")
display(museum_ambiguous_numeric)

print("[미술관] 95% 이상 100% 미만 숫자형 변환 후보의 비숫자 값")
display(art_ambiguous_numeric)
```

```python
# 숫자형 변환 대상 1차 분류
# 1) safe: 결측이 아닌 값 기준 100% numeric 변환 가능
# 2) ambiguous: 95% 이상 변환 가능하지만 일부 문자값 존재
# 3) hold: 95% 미만으로 변환 보류

def split_numeric_columns(check_df):
    safe_cols = check_df[
        check_df["numeric_convertible_rate"] == 100
    ]["column"].tolist()
    
    ambiguous_cols = check_df[
        (check_df["numeric_convertible_rate"] >= 95) &
        (check_df["numeric_convertible_rate"] < 100)
    ]["column"].tolist()
    
    hold_cols = check_df[
        check_df["numeric_convertible_rate"] < 95
    ]["column"].tolist()
    
    return safe_cols, ambiguous_cols, hold_cols


museum_numeric_safe_cols, museum_numeric_ambiguous_cols, museum_numeric_hold_cols = split_numeric_columns(museum_numeric_check)
art_numeric_safe_cols, art_numeric_ambiguous_cols, art_numeric_hold_cols = split_numeric_columns(art_numeric_check)

print("[박물관] 바로 변환 가능 컬럼 수:", len(museum_numeric_safe_cols))
print(museum_numeric_safe_cols)

print("\n[박물관] 일부 문자값 확인 필요 컬럼 수:", len(museum_numeric_ambiguous_cols))
print(museum_numeric_ambiguous_cols)

print("\n[박물관] 변환 보류 컬럼 수:", len(museum_numeric_hold_cols))
print(museum_numeric_hold_cols)

print("\n" + "=" * 80 + "\n")

print("[미술관] 바로 변환 가능 컬럼 수:", len(art_numeric_safe_cols))
print(art_numeric_safe_cols)

print("\n[미술관] 일부 문자값 확인 필요 컬럼 수:", len(art_numeric_ambiguous_cols))
print(art_numeric_ambiguous_cols)

print("\n[미술관] 변환 보류 컬럼 수:", len(art_numeric_hold_cols))
print(art_numeric_hold_cols)
```

```python
# 숫자형 변환 전 dtype 저장
# safe 컬럼만 먼저 변환하고, 변환 전후 dtype을 비교한다.

museum_dtype_before_numeric = museum_typed[museum_numeric_safe_cols].dtypes.astype(str)
art_dtype_before_numeric = art_typed[art_numeric_safe_cols].dtypes.astype(str)

print("[박물관] 숫자형 변환 전 dtype")
display(museum_dtype_before_numeric.to_frame("dtype_before"))

print("[미술관] 숫자형 변환 전 dtype")
display(art_dtype_before_numeric.to_frame("dtype_before"))
```

```python
# 숫자형 변환 함수
# 쉼표 제거 후 numeric 변환
# errors="coerce"를 사용하지만, safe 컬럼은 사전 점검에서 100% 변환 가능했던 컬럼이다.

def convert_to_numeric(df, cols):
    df = df.copy()
    
    for col in cols:
        df[col] = (
            df[col]
            .astype("string")
            .str.strip()
            .str.replace(",", "", regex=False)
        )
        df[col] = pd.to_numeric(df[col], errors="coerce")
    
    return df


museum_typed = convert_to_numeric(museum_typed, museum_numeric_safe_cols)
art_typed = convert_to_numeric(art_typed, art_numeric_safe_cols)

print("safe 컬럼 숫자형 변환 완료")
print("박물관 변환 컬럼 수:", len(museum_numeric_safe_cols))
print("미술관 변환 컬럼 수:", len(art_numeric_safe_cols))
```

```python
# 숫자형 변환 후 dtype 저장
museum_dtype_after_numeric = museum_typed[museum_numeric_safe_cols].dtypes.astype(str)
art_dtype_after_numeric = art_typed[art_numeric_safe_cols].dtypes.astype(str)

museum_numeric_dtype_compare = pd.DataFrame({
    "dtype_before": museum_dtype_before_numeric,
    "dtype_after": museum_dtype_after_numeric
})

art_numeric_dtype_compare = pd.DataFrame({
    "dtype_before": art_dtype_before_numeric,
    "dtype_after": art_dtype_after_numeric
})

print("[박물관] 숫자형 변환 전후 dtype 비교")
display(museum_numeric_dtype_compare)

print("[미술관] 숫자형 변환 전후 dtype 비교")
display(art_numeric_dtype_compare)
```

```python
# 숫자형 변환 과정에서 결측치가 새로 생겼는지 확인
# safe 컬럼은 원칙적으로 새 결측치가 생기지 않아야 한다.

def check_new_missing_after_numeric_conversion(df_original, df_converted, cols):
    rows = []
    
    for col in cols:
        before_missing = df_original[col].isna().sum()
        after_missing = df_converted[col].isna().sum()
        
        rows.append({
            "column": col,
            "missing_before": before_missing,
            "missing_after": after_missing,
            "new_missing_count": after_missing - before_missing
        })
    
    return pd.DataFrame(rows)


museum_numeric_missing_check = check_new_missing_after_numeric_conversion(
    museum_basic,
    museum_typed,
    museum_numeric_safe_cols
)

art_numeric_missing_check = check_new_missing_after_numeric_conversion(
    art_basic,
    art_typed,
    art_numeric_safe_cols
)

print("[박물관] 숫자형 변환 후 신규 결측 발생 확인")
display(
    museum_numeric_missing_check[
        museum_numeric_missing_check["new_missing_count"] != 0
    ]
)

print("[미술관] 숫자형 변환 후 신규 결측 발생 확인")
display(
    art_numeric_missing_check[
        art_numeric_missing_check["new_missing_count"] != 0
    ]
)
```

- 미술관 변환 컬럼 수: 24개
- dtype_after가 Float64로 변환

```python
# 날짜형 변환 가능 여부 확인
# 실제 변환은 아직 하지 않고, pd.to_datetime으로 변환 가능한 비율만 확인한다.

def date_conversion_check(df, cols, dataset_name):
    rows = []
    
    for col in cols:
        s = df[col]
        
        cleaned = (
            s.astype("string")
             .str.strip()
        )
        
        # 빈 문자열, '-', '미상' 등은 결측으로 간주
        cleaned = cleaned.replace({
            "": pd.NA,
            "-": pd.NA,
            "미상": pd.NA,
            "미정": pd.NA,
            "nan": pd.NA,
            "None": pd.NA
        })
        
        converted = pd.to_datetime(cleaned, errors="coerce")
        
        non_missing_count = cleaned.notna().sum()
        convertible_count = converted.notna().sum()
        
        rows.append({
            "dataset": dataset_name,
            "column": col,
            "dtype_before": str(s.dtype),
            "non_missing_count": non_missing_count,
            "date_convertible_count": convertible_count,
            "date_convertible_rate": round((convertible_count / non_missing_count) * 100, 2) if non_missing_count > 0 else 0,
            "sample_values": " | ".join(s.dropna().astype(str).unique()[:10])
        })
    
    return pd.DataFrame(rows)


museum_date_check = date_conversion_check(
    museum_typed,
    museum_date_cols,
    "박물관"
)

art_date_check = date_conversion_check(
    art_typed,
    art_date_cols,
    "미술관"
)

print("[박물관] 날짜형 변환 가능 여부")
display(museum_date_check)

print("[미술관] 날짜형 변환 가능 여부")
display(art_date_check)
```

```python
# 날짜형 변환 전 원본 보존 및 dtype 저장
# OPNNG_DE는 값 형태가 다양하므로 원본 컬럼을 별도로 보존한다.

date_cols_to_convert = ["BASE_DE", "LAST_CHG_DE", "OPNNG_DE"]

museum_date_convert_cols = [col for col in date_cols_to_convert if col in museum_typed.columns]
art_date_convert_cols = [col for col in date_cols_to_convert if col in art_typed.columns]

# OPNNG_DE 원본 보존
for df in [museum_typed, art_typed]:
    if "OPNNG_DE" in df.columns and "OPNNG_DE_RAW" not in df.columns:
        insert_idx = df.columns.get_loc("OPNNG_DE") + 1
        df.insert(insert_idx, "OPNNG_DE_RAW", df["OPNNG_DE"])

museum_dtype_before_date = museum_typed[museum_date_convert_cols].dtypes.astype(str)
art_dtype_before_date = art_typed[art_date_convert_cols].dtypes.astype(str)

print("[박물관] 날짜형 변환 전 dtype")
display(museum_dtype_before_date.to_frame("dtype_before"))

print("[미술관] 날짜형 변환 전 dtype")
display(art_dtype_before_date.to_frame("dtype_before"))
```

```python
# 날짜형 변환 함수 정의
# BASE_DE, LAST_CHG_DE는 일반 날짜 변환
# OPNNG_DE는 날짜 패턴을 추출한 뒤 변환한다.

import re

def normalize_missing_date_values(s):
    return (
        s.astype("string")
         .str.strip()
         .replace({
             "": pd.NA,
             "-": pd.NA,
             "미상": pd.NA,
             "미정": pd.NA,
             "nan": pd.NA,
             "None": pd.NA
         })
    )


def convert_general_date_series(s):
    cleaned = normalize_missing_date_values(s)
    return pd.to_datetime(cleaned, errors="coerce")


def parse_opening_date_value(value):
    if pd.isna(value):
        return pd.Series({
            "parsed_date": pd.NaT,
            "precision": pd.NA
        })
    
    text = str(value).strip()
    
    if text in ["", "-", "미상", "미정", "nan", "None"]:
        return pd.Series({
            "parsed_date": pd.NaT,
            "precision": pd.NA
        })
    
    # 한글 날짜 표현도 일부 대응
    text = (
        text.replace("년", ".")
            .replace("월", ".")
            .replace("일", "")
    )
    
    # YYYY / YYYY.MM / YYYY.MM.DD 형태 추출
    pattern = r"(\d{4})(?:[.\-/]\s*(\d{1,2}))?(?:[.\-/]\s*(\d{1,2}))?"
    
    for match in re.finditer(pattern, text):
        year = int(match.group(1))
        month = match.group(2)
        day = match.group(3)
        
        # 너무 비현실적인 연도는 제외
        if year < 1800 or year > 2100:
            continue
        
        if month is None:
            month = 1
            day = 1
            precision = "year"
        elif day is None:
            month = int(month)
            day = 1
            precision = "month"
        else:
            month = int(month)
            day = int(day)
            precision = "day"
        
        if not (1 <= month <= 12):
            continue
        
        if not (1 <= day <= 31):
            continue
        
        parsed = pd.to_datetime(
            f"{year:04d}-{month:02d}-{day:02d}",
            errors="coerce"
        )
        
        if pd.notna(parsed):
            return pd.Series({
                "parsed_date": parsed,
                "precision": precision
            })
    
    return pd.Series({
        "parsed_date": pd.NaT,
        "precision": pd.NA
    })


def convert_date_columns(df):
    df = df.copy()
    
    # 일반 날짜 컬럼 변환
    for col in ["BASE_DE", "LAST_CHG_DE"]:
        if col in df.columns:
            df[col] = convert_general_date_series(df[col])
    
    # 개관일 변환
    if "OPNNG_DE" in df.columns:
        raw_col = "OPNNG_DE_RAW" if "OPNNG_DE_RAW" in df.columns else "OPNNG_DE"
        parsed_result = df[raw_col].apply(parse_opening_date_value)
        
        df["OPNNG_DE"] = parsed_result["parsed_date"]
        df["OPNNG_DE_PRECISION"] = parsed_result["precision"]
    
    return df


museum_typed = convert_date_columns(museum_typed)
art_typed = convert_date_columns(art_typed)

print("날짜형 변환 완료")
```

```python
# 날짜형 변환 검증
# OPNNG_DE_RAW는 원본, OPNNG_DE는 변환된 날짜
# OPNNG_DE_PRECISION은 날짜 정확도 수준이다.

def check_date_missing_after_conversion(df, cols):
    rows = []
    
    for col in cols:
        if col == "OPNNG_DE" and "OPNNG_DE_RAW" in df.columns:
            raw_s = normalize_missing_date_values(df["OPNNG_DE_RAW"])
        else:
            raw_s = normalize_missing_date_values(df[col])
        
        before_missing = raw_s.isna().sum()
        after_missing = df[col].isna().sum()
        
        rows.append({
            "column": col,
            "missing_before": before_missing,
            "missing_after": after_missing,
            "new_missing_count": after_missing - before_missing
        })
    
    return pd.DataFrame(rows)


museum_date_missing_check = check_date_missing_after_conversion(
    museum_typed,
    museum_date_convert_cols
)

art_date_missing_check = check_date_missing_after_conversion(
    art_typed,
    art_date_convert_cols
)

print("[박물관] 날짜형 변환 후 신규 결측 확인")
display(museum_date_missing_check)

print("[미술관] 날짜형 변환 후 신규 결측 확인")
display(art_date_missing_check)

print("[박물관] OPNNG_DE 날짜 정확도 수준")
display(museum_typed["OPNNG_DE_PRECISION"].value_counts(dropna=False).to_frame("count"))

print("[미술관] OPNNG_DE 날짜 정확도 수준")
display(art_typed["OPNNG_DE_PRECISION"].value_counts(dropna=False).to_frame("count"))
```

### 03. 자료형 변환 단계 정리

##### 1) 숫자형 변환

- 쉼표가 포함된 숫자 문자열은 쉼표를 제거한 뒤 `numeric` 자료형으로 변환하였다.
- 숫자형 변환 가능 비율이 100%인 컬럼을 우선 변환하였다.
- 일부 비숫자 값이 섞여 있더라도 컬럼 의미상 숫자형으로 해석하는 것이 적절한 컬럼은 숫자형으로 변환하였다.
- 숫자로 해석하기 어려운 값은 `NaN`으로 처리되었다.
- 설명형 값, 복합 가격 정보, 문자 혼합 값이 많은 컬럼은 무리하게 변환하지 않고 보류하였다.

##### 2) 날짜형 변환

- `BASE_DE`, `LAST_CHG_DE`, `OPNNG_DE` 컬럼을 날짜형으로 변환하였다.
- `OPNNG_DE`는 개관일 정보이지만 연도만 있는 값, 연월만 있는 값, 복수 날짜가 포함된 값, 설명이 함께 적힌 값이 존재하였다.
- 따라서 `OPNNG_DE`의 원본 값은 `OPNNG_DE_RAW` 컬럼으로 보존하였다.
- 날짜로 해석 가능한 첫 번째 날짜 패턴을 추출하여 `OPNNG_DE`를 변환하였다.
- 날짜 정확도 수준을 구분하기 위해 `OPNNG_DE_PRECISION` 컬럼을 추가하였다.
  - `day`: 연월일 수준
  - `month`: 연월 수준
  - `year`: 연도 수준

##### 3) 변환 결과

- 박물관 데이터의 숫자형 변환 대상 컬럼은 정상적으로 변환되었다.
- 미술관 데이터의 숫자형 변환 대상 컬럼도 정상적으로 변환되었다.
- 날짜형 컬럼은 모두 `datetime` 자료형으로 변환되었다.
- 안전 변환 컬럼에서는 신규 결측치가 발생하지 않았다.
- 일부 문자값이 섞인 컬럼에서는 숫자로 해석할 수 없는 값이 `NaN`으로 처리되었다.

##### 4) 주의 사항

- 현재 단계에서는 자료형 변환만 진행하였다.
- 결측치 대체, 이상치 처리, 행 제거는 아직 추가로 진행하지 않았다.
- `DAY_AVRG_VIEWNG_NMPR_CO`는 숫자형으로 변환했지만, 이후 관람객 수 예측 모델에서 데이터 누수 가능성이 있으므로 입력 변수로 사용할 때 주의가 필요하다.

```python
# 시설 유형별 결측치 비교
# 같은 컬럼이라도 박물관/미술관별 결측 양상이 다를 수 있으므로 분리해서 확인한다.

def missing_summary_by_facility_type(df, group_col="facility_type"):
    rows = []
    
    for facility_type, group_df in df.groupby(group_col):
        for col in group_df.columns:
            if col == group_col:
                continue
            
            rows.append({
                "facility_type": facility_type,
                "column": col,
                "missing_count": group_df[col].isna().sum(),
                "missing_rate": round(group_df[col].isna().mean() * 100, 2),
                "non_missing_count": group_df[col].notna().sum(),
                "dtype": str(group_df[col].dtype),
                "unique_count": group_df[col].nunique(dropna=True)
            })
    
    return pd.DataFrame(rows)


facility_missing_by_type = missing_summary_by_facility_type(facility_common)

print("[시설 유형별 결측치 비교]")
display(
    facility_missing_by_type.sort_values(
        by=["column", "facility_type"]
    )
)
```

```python
# 06.2 좌표 컬럼 범위 검증
# 시설 위치 좌표가 국내 시설 범위에서 크게 벗어나는지 확인한다.
# 일반적으로 한국 경도는 약 124~132, 위도는 약 33~39 범위로 볼 수 있다.

coordinate_check = pd.DataFrame([
    {
        "column": "FCLTY_LO",
        "expected_range": "124 ~ 132",
        "min": facility_common["FCLTY_LO"].min(),
        "max": facility_common["FCLTY_LO"].max(),
        "out_of_range_count": (
            (facility_common["FCLTY_LO"] < 124) |
            (facility_common["FCLTY_LO"] > 132)
        ).sum()
    },
    {
        "column": "FCLTY_LA",
        "expected_range": "33 ~ 39",
        "min": facility_common["FCLTY_LA"].min(),
        "max": facility_common["FCLTY_LA"].max(),
        "out_of_range_count": (
            (facility_common["FCLTY_LA"] < 33) |
            (facility_common["FCLTY_LA"] > 39)
        ).sum()
    }
])

display(coordinate_check)
```

```python
# 06.3 0 값 확인
# 0이 실제 의미 있는 값인지, 입력 오류인지 판단하기 위해 주요 컬럼의 0 값 행을 확인한다.

zero_check_cols = [
    "OPNNG_DAY_CO",
    "NMPR_CO",
    "QUALF_HOLD_CO",
    "ARTGR_EMP_CO",
    "TOT_PROGRM_CO"
]

zero_check_cols = [
    col for col in zero_check_cols
    if col in facility_common.columns
]

for col in zero_check_cols:
    zero_rows = facility_common[
        facility_common[col] == 0
    ][[
        "facility_type", "ID", "FCLTY_NM", "CTPRVN_NM", "SIGNGU_NM", col
    ]]
    
    print(f"[{col}] 0 값 행 수:", len(zero_rows))
    display(zero_rows.head(20))
```

```python
# 06.4 IQR 기준 이상치 후보 요약
# 이상치를 제거하지 않고, 후보 개수와 비율만 확인한다.
# 좌표 컬럼은 별도 범위 검증을 했으므로 IQR 이상치 확인 대상에서 제외한다.

iqr_outlier_cols = [
    col for col in numeric_cols_for_outlier
    if col not in ["FCLTY_LO", "FCLTY_LA"]
]

def iqr_outlier_summary(df, cols):
    rows = []
    
    for col in cols:
        s = df[col].dropna()
        
        if len(s) == 0:
            continue
        
        q1 = s.quantile(0.25)
        q3 = s.quantile(0.75)
        iqr = q3 - q1
        
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        
        outlier_mask = (df[col] < lower_bound) | (df[col] > upper_bound)
        outlier_count = outlier_mask.sum()
        non_missing_count = df[col].notna().sum()
        
        rows.append({
            "column": col,
            "non_missing_count": non_missing_count,
            "q1": q1,
            "median": s.median(),
            "q3": q3,
            "iqr": iqr,
            "lower_bound": lower_bound,
            "upper_bound": upper_bound,
            "min": s.min(),
            "max": s.max(),
            "outlier_count": outlier_count,
            "outlier_rate": round(outlier_count / non_missing_count * 100, 2)
        })
    
    return pd.DataFrame(rows).sort_values(
        by="outlier_rate",
        ascending=False
    )


iqr_outlier_result = iqr_outlier_summary(
    facility_common,
    iqr_outlier_cols
)

display(iqr_outlier_result)
```

```python
# 06.5 주요 이상치 후보 상세 확인
# IQR 기준 이상치 후보가 많은 컬럼 중 분석적으로 중요한 컬럼을 우선 확인한다.
# 여기서도 제거하지 않고, 어떤 시설이 극단값을 가지는지만 확인한다.

important_outlier_cols = [
    "LND_AR_VALUE",
    "EDC_FCLTY_AR_VALUE",
    "DATA_SPCE_AR_VALUE",
    "NMPR_CO",
    "PUBLIC_VLNTER_CO",
    "ARTGR_EMP_CO",
    "DATA_CO",
    "TOT_PROGRM_CO",
    "VIEWNG_NMPR_CO",
    "DAY_AVRG_VIEWNG_NMPR_CO"
]

important_outlier_cols = [
    col for col in important_outlier_cols
    if col in facility_common.columns
]

def show_top_values(df, col, n=10):
    check_cols = [
        "facility_type", "ID", "FCLTY_NM", "CTPRVN_NM", "SIGNGU_NM", col
    ]
    
    print(f"[{col}] 상위 {n}개 값")
    display(
        df[check_cols]
        .dropna(subset=[col])
        .sort_values(col, ascending=False)
        .head(n)
    )


for col in important_outlier_cols:
    show_top_values(facility_common, col, n=10)
```

- QUALF_HOLD_CO 0값 240개 → 자격 보유 인력 없음으로 해석 가능

IQR 이상치:

- 하지만 대부분 규모가 큰 정상 시설일 가능성이 높음
- 따라서 현재 단계에서는 제거하지 않고, 이상치 플래그만 만드는 방향이 적절함

```python
# 06.6 이상치 처리 방향 1차 정리
# 현재 단계에서는 이상치를 제거하지 않고, 처리 방향만 정리한다.

outlier_decision_summary = pd.DataFrame([
    {
        "category": "좌표",
        "columns": "FCLTY_LO, FCLTY_LA",
        "finding": "국내 시설 좌표 범위에서 벗어나는 값 없음",
        "decision": "유지"
    },
    {
        "category": "0 값",
        "columns": "OPNNG_DAY_CO",
        "finding": "0 값은 대부분 휴관/폐관 시설에서 발생",
        "decision": "유지하되, 운영일수 분석 시 별도 확인"
    },
    {
        "category": "0 값",
        "columns": "NMPR_CO, QUALF_HOLD_CO, ARTGR_EMP_CO, TOT_PROGRM_CO",
        "finding": "0 값이 존재하지만 실제 0명/0개일 가능성이 있음",
        "decision": "현재 단계에서는 유지"
    },
    {
        "category": "IQR 이상치",
        "columns": "LND_AR_VALUE, DATA_CO, VIEWNG_NMPR_CO 등",
        "finding": "규모가 큰 시설에서 큰 값이 관측됨",
        "decision": "제거하지 않고 플래그 생성"
    },
    {
        "category": "모델링 주의",
        "columns": "DAY_AVRG_VIEWNG_NMPR_CO",
        "finding": "관람객 수와 직접적으로 연결되는 파생 성과 변수",
        "decision": "EDA에서는 참고 가능하나, 관람객 수 예측 모델 입력 변수 사용 주의"
    }
])

display(outlier_decision_summary)
```

```python
# IQR 이상치 플래그 데이터 생성
# 원본 facility_common은 그대로 유지하고, 플래그가 추가된 별도 데이터셋을 생성한다.

facility_outlier_flagged = facility_common.copy()

outlier_flag_cols = []

for col in iqr_outlier_cols:
    s = facility_outlier_flagged[col].dropna()
    
    if len(s) == 0:
        continue
    
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    flag_col = f"{col}_IQR_OUTLIER"
    
    facility_outlier_flagged[flag_col] = (
        (facility_outlier_flagged[col] < lower_bound) |
        (facility_outlier_flagged[col] > upper_bound)
    )
    
    outlier_flag_cols.append(flag_col)

print("IQR 이상치 플래그 생성 완료")
print("생성된 플래그 컬럼 수:", len(outlier_flag_cols))
print(outlier_flag_cols)
```

```python
# 이상치 플래그별 True 개수 확인

outlier_flag_summary = []

for flag_col in outlier_flag_cols:
    original_col = flag_col.replace("_IQR_OUTLIER", "")
    
    outlier_flag_summary.append({
        "column": original_col,
        "flag_column": flag_col,
        "outlier_count": facility_outlier_flagged[flag_col].sum(),
        "outlier_rate": round(facility_outlier_flagged[flag_col].mean() * 100, 2)
    })

outlier_flag_summary = pd.DataFrame(outlier_flag_summary).sort_values(
    "outlier_rate",
    ascending=False
)

display(outlier_flag_summary)
```

```python
# 시설별로 몇 개 컬럼에서 IQR 이상치 후보로 잡혔는지 확인한다.
# 여러 컬럼에서 동시에 이상치로 잡히는 시설은 대형 시설일 가능성이 높다.

facility_outlier_flagged["IQR_OUTLIER_FLAG_COUNT"] = (
    facility_outlier_flagged[outlier_flag_cols].sum(axis=1)
)

print("[시설별 IQR 이상치 플래그 개수 상위 20개]")
display(
    facility_outlier_flagged[
        [
            "facility_type", "ID", "FCLTY_NM", "CTPRVN_NM", "SIGNGU_NM",
            "IQR_OUTLIER_FLAG_COUNT"
        ]
    ]
    .sort_values("IQR_OUTLIER_FLAG_COUNT", ascending=False)
    .head(20)
)

print("[IQR 이상치 플래그 개수 분포]")
display(
    facility_outlier_flagged["IQR_OUTLIER_FLAG_COUNT"]
    .value_counts()
    .sort_index()
    .to_frame("facility_count")
)
```

```python
# 로그 변환 후보 컬럼 확인
# 오른쪽 꼬리가 긴 규모/관람객/자료 수 변수는 이후 EDA나 모델링에서 log1p 변환을 고려할 수 있다.

log_transform_candidate_cols = [
    "LND_AR_VALUE",
    "EDC_FCLTY_AR_VALUE",
    "DATA_SPCE_AR_VALUE",
    "NMPR_CO",
    "DATA_CO",
    "TOT_PROGRM_CO",
    "VIEWNG_NMPR_CO",
    "DAY_AVRG_VIEWNG_NMPR_CO"
]

log_transform_candidate_cols = [
    col for col in log_transform_candidate_cols
    if col in facility_common.columns
]

log_candidate_summary = []

for col in log_transform_candidate_cols:
    s = facility_common[col].dropna()
    
    log_candidate_summary.append({
        "column": col,
        "non_missing_count": s.count(),
        "min": s.min(),
        "median": s.median(),
        "max": s.max(),
        "skew": round(s.skew(), 3)
    })

log_candidate_summary = pd.DataFrame(log_candidate_summary).sort_values(
    "skew",
    ascending=False
)

display(log_candidate_summary)
```

### 06. 이상치 확인 결과

#### 1) 좌표값 확인

- 시설 위치 좌표인 `FCLTY_LO`, `FCLTY_LA`의 범위를 확인하였다.

- 국내 시설 좌표 범위를 크게 벗어나는 값은 확인되지 않았다.

#### 2) 음수 값 확인

- 숫자형 컬럼 전체에서 음수 값은 확인되지 않았다.
- 면적, 인력 수, 자료 수, 프로그램 수, 관람객 수 등에서 음수 값이 없으므로 기본적인 수치 입력 오류 가능성은 낮다.

#### 3) 0 값 확인

- `OPNNG_DAY_CO`에서 0 값이 12개 확인되었다.
- 해당 시설은 대부분 시설명에 휴관 또는 폐관 정보가 포함되어 있어, 단순 오류라기보다 운영 상태와 관련된 값으로 판단된다.
- `NMPR_CO`, `QUALF_HOLD_CO`, `ARTGR_EMP_CO`, `TOT_PROGRM_CO`에서도 0 값이 확인되었다.
- 이 값들은 실제 인력 없음 또는 프로그램 없음일 가능성이 있으므로 현재 단계에서는 제거하지 않는다.

#### 4) IQR 기준 이상치 확인

- IQR 기준으로 여러 숫자형 컬럼에서 이상치 후보가 확인되었다.
- 이상치 후보가 많은 컬럼은 `PUBLIC_VLNTER_CO`, `PRVATE_VLNTER_CO`, `QUALF_HOLD_CO`, `LND_AR_VALUE`, `VIEWNG_NMPR_CO`, `DAY_AVRG_VIEWNG_NMPR_CO`, `DATA_CO` 등이다.
- 그러나 상위 이상치 시설을 확인한 결과, 국립중앙박물관, 국립현대미술관, 독립기념관 등 대형 시설이 주로 포함되어 있었다.
- 따라서 해당 값들은 단순 오류라기보다 시설 규모 차이에서 발생한 정상적인 극단값일 가능성이 높다.

#### 5) 이상치 처리 방향

- 현재 단계에서는 이상치를 제거하지 않는다.
- 원본 통합 데이터인 `facility_common`은 그대로 유지한다.
- IQR 기준 이상치 여부를 나타내는 플래그 컬럼을 추가한 `facility_outlier_flagged` 데이터셋을 별도로 생성하였다.
- 시설별 이상치 플래그 개수인 `IQR_OUTLIER_FLAG_COUNT`를 생성하였다.
- 이후 EDA에서 대형 시설 여부를 해석하거나, 모델링 단계에서 이상치 영향도를 확인할 때 활용할 수 있다.

```python
# 로그 변환 전후 왜도 비교
# 왜도 감소 여부를 확인해 로그 변환의 효과를 검토한다.

log_skew_compare_rows = []

for col in log_transform_cols:
    original_s = facility_eda[col].dropna()
    log_s = facility_eda[f"LOG1P_{col}"].dropna()
    
    log_skew_compare_rows.append({
        "column": col,
        "original_skew": round(original_s.skew(), 3),
        "log1p_skew": round(log_s.skew(), 3),
        "skew_reduction": round(abs(original_s.skew()) - abs(log_s.skew()), 3)
    })

log_skew_compare = pd.DataFrame(log_skew_compare_rows).sort_values(
    "skew_reduction",
    ascending=False
)

display(log_skew_compare)
```

- `세종특별자치시`는 최초 매핑에서 누락되어 있었으나, 추가 보정하여 `충청권`으로 분류하였다.

권역별 시설 수는 다음과 같다.

| 권역 | 시설 수 |
|---|---:|
| 수도권 | 401 |
| 경상권 | 245 |
| 전라권 | 195 |
| 충청권 | 150 |
| 강원권 | 118 |
| 제주권 | 82 |

#### 4) 로그 변환 파생변수

- 오른쪽 꼬리가 긴 분포를 보이는 숫자형 변수에 대해 `log1p` 변환 컬럼을 생성하였다.
- 로그 변환은 0 값이 포함된 변수에도 적용할 수 있도록 `np.log1p()`를 사용하였다.
- 생성한 로그 변환 컬럼은 다음과 같다.
  - `LOG1P_LND_AR_VALUE`
  - `LOG1P_EDC_FCLTY_AR_VALUE`
  - `LOG1P_DATA_SPCE_AR_VALUE`
  - `LOG1P_NMPR_CO`
  - `LOG1P_DATA_CO`
  - `LOG1P_TOT_PROGRM_CO`
  - `LOG1P_VIEWNG_NMPR_CO`
  - `LOG1P_DAY_AVRG_VIEWNG_NMPR_CO`

#### 5) 로그 변환 효과

- 로그 변환 후 주요 규모 변수와 관람객 수 변수의 왜도가 크게 감소하였다.
- 특히 `LND_AR_VALUE`, `DATA_SPCE_AR_VALUE`, `VIEWNG_NMPR_CO`, `DAY_AVRG_VIEWNG_NMPR_CO`에서 로그 변환 효과가 뚜렷하게 나타났다.
- 따라서 이후 EDA에서는 원본 변수와 로그 변환 변수를 함께 비교할 수 있다.
- 모델링 단계에서는 분포 왜곡이 큰 변수에 대해 로그 변환 변수 사용을 고려할 수 있다.

#### 6) 주의 사항

- `DAY_AVRG_VIEWNG_NMPR_CO`와 `LOG1P_DAY_AVRG_VIEWNG_NMPR_CO`는 관람객 수와 직접적으로 연결된 성과 변수이다.
- 이후 관람객 수 예측 모델에서는 데이터 누수 가능성이 있으므로 입력 변수로 사용할 때 주의가 필요하다.
- 현재 단계에서는 파생변수 생성까지만 진행했으며, 최종 분석용 변수 선택은 이후 EDA와 모델링 단계에서 다시 판단한다.

### 1. 프로젝트 개요

### 6. 자료형 변환

숫자형과 날짜형 컬럼을 구분하여 자료형 변환을 진행하였다.

#### 숫자형 변환

- 쉼표가 포함된 숫자 문자열은 쉼표를 제거한 뒤 숫자형으로 변환하였다.

- 숫자형 변환 가능 비율이 100%인 컬럼을 우선 변환하였다.
- 일부 비숫자 값이 섞여 있더라도 컬럼 의미상 숫자형으로 보는 것이 적절한 컬럼은 숫자형으로 변환하였다.
- 설명형 값, 복합 가격 정보, 문자 혼합 값이 많은 컬럼은 무리하게 변환하지 않고 보류하였다.

`OPNNG_DE_PRECISION`의 의미는 다음과 같다.

### 10. 이상치 확인

숫자형 컬럼을 대상으로 이상치 구조를 확인하였다.

#### 좌표값 확인

- `FCLTY_LO`, `FCLTY_LA`의 범위를 확인하였다.
- 국내 시설 좌표 범위를 크게 벗어나는 값은 확인되지 않았다.

#### 음수 값 확인

- 숫자형 컬럼 전체에서 음수 값은 확인되지 않았다.

#### 0 값 확인

- `OPNNG_DAY_CO`에서 0 값이 일부 확인되었다.
- 대부분 시설명에 휴관 또는 폐관 정보가 포함되어 있어 단순 오류라기보다 운영 상태와 관련된 값으로 판단하였다.
- 인력 수, 프로그램 수 관련 컬럼에서도 0 값이 확인되었지만 실제 0명 또는 0개일 가능성이 있으므로 현재 단계에서는 유지하였다.

#### IQR 기준 이상치 확인

- IQR 기준으로 여러 숫자형 컬럼에서 이상치 후보가 확인되었다.
- 그러나 상위 이상치 시설을 확인한 결과, 국립중앙박물관, 국립현대미술관, 독립기념관 등 대형 시설이 주로 포함되어 있었다.
- 따라서 해당 값들은 단순 오류라기보다 시설 규모 차이에서 발생한 정상적인 극단값일 가능성이 높다고 판단하였다.

현재 단계에서는 이상치를 제거하지 않고, IQR 기준 이상치 여부를 나타내는 플래그 컬럼을 생성하였다.  
또한 시설별 이상치 플래그 개수인 `IQR_OUTLIER_FLAG_COUNT`를 생성하였다.

### 12. EDA 진행 시 주의 사항

- `facility_common`은 박물관/미술관 통합 기본 데이터이다.
- `facility_eda`는 이상치 플래그와 파생변수가 포함된 EDA용 데이터이다.
- 이후 EDA는 `facility_eda`를 기준으로 진행한다.
- 현재 단계에서는 결측치 대체, 이상치 제거, 모델링용 변수 선택은 아직 진행하지 않았다.
- `DAY_AVRG_VIEWNG_NMPR_CO`와 `LOG1P_DAY_AVRG_VIEWNG_NMPR_CO`는 관람객 수와 직접적으로 연결된 성과 변수이다.
- 이후 관람객 수 예측 모델에서는 데이터 누수 가능성이 있으므로 입력 변수로 사용할 때 주의가 필요하다.

### 13. 최종 산출물

EDA를 별도 노트북에서 진행하기 위해 전처리 완료 데이터를 CSV 파일로 저장한다.

저장 대상 데이터는 다음과 같다.

- `facility_eda`: EDA용 최종 전처리 데이터
- 필요 시 `facility_common`: 파생변수 생성 전 통합 기본 데이터

다음 단계에서는 저장된 CSV 파일을 새 EDA 노트북에서 불러와 시설 유형별, 권역별, 시도별 기초 분석부터 진행한다.

```python
# 전처리 완료 데이터 CSV 저장
# EDA는 별도 노트북에서 진행할 예정이므로, 현재까지 전처리한 데이터를 CSV로 저장한다.

import os

processed_dir = "data/processed"
os.makedirs(processed_dir, exist_ok=True)

facility_eda_path = os.path.join(processed_dir, "facility_eda_preprocessed.csv")
facility_common_path = os.path.join(processed_dir, "facility_common_base.csv")

# EDA용 최종 데이터 저장
facility_eda.to_csv(
    facility_eda_path,
    index=False,
    encoding="utf-8-sig",
    date_format="%Y-%m-%d"
)

# 파생변수 생성 전 통합 기본 데이터도 함께 저장
facility_common.to_csv(
    facility_common_path,
    index=False,
    encoding="utf-8-sig",
    date_format="%Y-%m-%d"
)

print("CSV 저장 완료")
print("EDA용 전처리 데이터:", facility_eda_path)
print("통합 기본 데이터:", facility_common_path)
```

```python
# 저장된 CSV 파일이 정상적으로 생성되었는지 확인한다.

saved_files = pd.DataFrame([
    {
        "file_name": "facility_eda_preprocessed.csv",
        "path": facility_eda_path,
        "exists": os.path.exists(facility_eda_path),
        "file_size_mb": round(os.path.getsize(facility_eda_path) / (1024 ** 2), 3) if os.path.exists(facility_eda_path) else None
    },
    {
        "file_name": "facility_common_base.csv",
        "path": facility_common_path,
        "exists": os.path.exists(facility_common_path),
        "file_size_mb": round(os.path.getsize(facility_common_path) / (1024 ** 2), 3) if os.path.exists(facility_common_path) else None
    }
])

display(saved_files)
```
